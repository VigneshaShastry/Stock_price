```python
#df_google_new.plot(x="date", y="close", figsize=(16,8))
#plt.show()
df_google_new=df_google.close.tolist()
df_google_new
```

```python

window_size=100
average=[]
i=0
for i in range(len(df_google)-window_size):
    df_google_split=df_google_new[i:window_size+i]
    temp=sum(df_google_split)/window_size
    average.append(temp)
last_n_list = df_google_new[-window_size:]
temp2=sum(df_google_split)/window_size
g_list=[temp2] * window_size  
moving_average=average+g_list
```

```python
df_google["Moving_Average"]=moving_average
df_google.plot(x="date", y="Moving_Average", figsize=(16,8))
plt.show()
```
